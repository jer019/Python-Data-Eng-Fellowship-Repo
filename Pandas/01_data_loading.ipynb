## Data Loading
pd.read_csv(): Read a CSV file into a DataFrame.  
pd.read_excel(): Read an Excel file into a DataFrame.  
pd.read_sql(): Read SQL query or database table into a DataFrame.  
pd.read_json(): Read a JSON file into a DataFrame.  
pd.read_html(): Read HTML tables into a list of DataFrame objects.  
pd.read_clipboard(): Read text from the clipboard into a DataFrame.
